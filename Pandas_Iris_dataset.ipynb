```python
df=iris.copy()
df.columns=["sl","sw","pl","pw","flower_type"]
df.head()
```

```python
df.describe()
```

```python
#first method drop na rows
df.dropna(inplace=True)
```

```python
df.head()
df.reset_index(drop=True,inplace=True)
df.head()
```

```python
df.iloc[2:4,1:3]=np.nan
df.head()
```

```python
#second method to fill na with mean of that column
df.sw.fillna(df.sw.mean(),inplace=True)
df.pl.fillna(df.pl.mean(),inplace=True)
df.head()
```

```python
#string based data
df["gender"]="female"
df.iloc[0:10,5]="male"
df.head()
```

```python
def f(s):
    if s=="male":
        return 1
    else:
        return 0
df["sex"]=df.gender.apply(f)
del df["gender"]
df.head()
```
